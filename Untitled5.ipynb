```
import requests
from pprint import pprint
import os
import pandas as pd
```

```
subscription_key = "8e63222ae4ac4591b2d69858c014a701"
endpoint = "https://textanalytics123456.cognitiveservices.azure.com/"
language_api_url = endpoint + "/text/analytics/v2.1/languages"
sentiment_url = endpoint + "/text/analytics/v2.1/sentiment"
keyphrase_url = endpoint + "/text/analytics/v2.1/keyphrases"
headers = {"Ocp-Apim-Subscription-Key": subscription_key}
```

```
def extract_language(ids,text):
    doc={}
    doc_lis={}
    doc_lis["id"]=ids
    doc_lis["text"]=text
    lis=[]
    lis.append(doc_lis)
    doc['documents']=lis
    #print(doc)
    response = requests.post(language_api_url, headers=headers, json=doc)
    languages = response.json()
    lan=''
    for j in languages['documents']:
        for k in j['detectedLanguages']:
            lan=k['iso6391Name']
    return lan

def sentiment_analysis(ids,lan,text):
    doc1={}
    doc_lis1={}
    doc_lis1["id"]=ids
    doc_lis1["text"]=text
    doc_lis1["language"]=lan
    lis1=[]
    lis1.append(doc_lis1)
    doc1['documents']=lis1
    #print(doc1)
    response = requests.post(sentiment_url, headers=headers, json=doc1)
    sentiments = response.json()
    return sentiments

def extract_keyword(ids,lang,text):
    doc1={}
    doc_lis1={}
    doc_lis1["id"]=ids
    doc_lis1["text"]=text
    doc_lis1["language"]=lan
    lis1=[]
    lis1.append(doc_lis1)
    doc1['documents']=lis1
    response = requests.post(keyphrase_url, headers=headers, json=doc1)
    key_phrases = response.json()
    return key_phrases
```
